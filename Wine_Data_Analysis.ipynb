Com o banco carregado, é interessante fazer uma rápida descrição dos dados:


*   **Acidez_fixa:** ácidos que não são capazes de se volatilizar (ácidos fixos), como exemplo, ácido tartárico, málico e cítrico, provenientes da uva, e lático e succínico, provenientes da fermentação
*   **Volatile acidity:** acidez e aroma muito desagradável, provocado pela presença de teor elevado de ácido acético e de acetato de etila no vinho, compostos produzidos essencialmente por bactérias acéticas.
*   **Citric acid:** é responsável pela sensação de frescor da bebida.
*   **Residual sugar:** As uvas têm dois principais tipos de açúcares, a glicose e a frutose (duas vezes mais doce que a glicose). As duas podem sobrar na bebida e serão chamadas de açúcar residual, se restar mais frutose nesse residual o vinho terá a sensação de maior dulçor.
*   **Chlorides:** é uma substancia presente nos vinhos
*   **Free sulfur dioxide:** é utilizado como conservante, devido às suas propriedades antioxidantes e antibacterianas, evitando que os vinhos degradem de forma mais rápida.
*   **Total sulfur dioxide:** é um composto muito importante utilizado na conservação de vinhos, devido às suas propriedades antioxidantes.
*   **Density:** está relacionada principalmente ao seu teor alcoólico e de açúcares residuais.
*   **pH:** O pH (potencial de hidrogênio) do mosto do vinho é importante pois, ele influencia na microbiologia e como vai ser gerenciado o uso de sulfito para evitar a oxidação do mesmo. Quanto mais alto o PH, mais sulfito é necessário.
*   **Sulphates:** sulfito é o nome popular utilizado para referir-se ao dióxido de enxofre (SO2)
*   **Alcohol:** O teor alcoólico é igual ao número de litros de álcool etílico contidos em 100 litros de vinho, sendo os dois volumes medidos a 20° C. 
*   **QUality:** é a qualidade dos vinhos.

```python
y = df["quality"] # Resposta
x = df.iloc[:, 0:11] # Dados de base
```

Na questão 3, foi preciso criar uma lista que sorteasse 3 números diferentes utilizando um laço "for", essa lista seria necessária para a realização das próximas questões da prova.

```python
lista = [] # Criando uma lista vazia

for numeros in range(3): # Limitando o laço for como 3 voltas
    lista.append(np.random.randint(11)) # Gerando número aleatório de 0 a 11
print(lista) # Mostrando a lista
```

Na questão 4, foi necessário utilizar a lista gerada na célula anterior como iterador no meu conjunto de dados.

```python
for i in lista: # Criando um "for" para o loop em cima da lista
    print(f"5 primeiras linhas da coluna {x.columns[i]}:")
    print(x.iloc[:, i].head()) # 5 primeiras linhas das colunas da lista
    print("")

    print(f"5 últimas linhas da coluna {x.columns[i]}:")
    print(x.iloc[:, i].tail()) # 5 últimas linhas das colunas da lista
    print("")

    print(f"Menor valor de cada coluna {x.columns[i]}")
    print(x.iloc[:, i].min()) # Menor valor de cada coluna da lista
    print("")

    print(f"Maior valor de cada coluna {x.columns[i]}")
    print(x.iloc[:, i].max()) # Maior valor de cada coluna da lista
    print("")

    print(f"Desvio padrão de cada coluna {x.columns[i]}")
    print(x.iloc[:, i].std()) # Desvio padrão de cada coluna da lista
    print("")
    
    print(f"Variancia de cada coluna {x.columns[i]}")
    print(x.iloc[:, i].var()) #  Variancia de cada coluna da lista
    print("")
    print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-")
    print("")
```

Na questão 6, era solicitado a visualização dos dados utilizando a biblioteca Matplotlib para fazer o plot das colunas sorteadas, através de um processo de iteração.

```python
for i in lista:
    plt.plot(x.iloc[:, i], 'blue') # Plota o gráfico na cor azul
    plt.rc('axes', titlesize=18)     # Tamanho da fonte no título
    plt.rc('axes', labelsize=10)    # Tamanho da fonte nos eixos x e y
    plt.rc('xtick', labelsize=13)    # Tamanho da fonte dos valores do eixo x
    plt.rc('ytick', labelsize=13)    # Tamanho da fonte dos valores do eixo y
    plt.rc('legend', fontsize=13)    # Tamanho da fonte da legenda
    # plt.rc('font', size=14) # Controla o tamanho default das fontes
    colname = x.columns[i] # pega o nome da coluna
    plt.xlabel(colname) # adiciona um label no eixo x
    plt.ylabel("Valor") # adiciona um label no eixo y
    plt.title(f"Gráfico do {colname}") # adiciona um título no gráfico
    plt.show() # mostra o gráfico
    print("")
```
